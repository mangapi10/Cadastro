```
nome =input(" Digite seu nome: ").strip().title()
sobrenome =input(" Digite seu sobrenome: ").strip().title()
salário=float(input(" Digite seu salário: "))
salario1=salário+salário*0.1
salario2=salário+salário*0.2
salario3=salário+salário*0.3
salario4=salário+salário*0.4
salario5=salário+salário*0.5

print (f'olá, {nome} {sobrenome}')
print (f'seu salário é de R${salário:.2f}')
print (f'seu salario com 10% de aumento é de R${salario1:.2f}')
print (f'seu salario com 20% de aumento é de R${salario2:.2f}')
print (f'seu salario com 30% de aumento é de R${salario3:.2f}')
print (f'seu salario com 40% de aumento é de R${salario4:.2f}')
print (f'seu salario com 50€ de aumento é de R${salario5:.2f}')
```
